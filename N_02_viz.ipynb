```python
# STATTER PLOT
# BASIC

def basic_scatterplot(data,x,y,title,xlabel,ylabel):
    sns.scatterplot(data = data, x = x, y = y, color = "purple")
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()


data = sns.load_dataset("tips")

basic_scatterplot(data,"tip","total_bill","Propina vs. Total pagado","Propina","Total pagado")
```

```python
# STATTER PLOT
# MEDIUM

def medium_scatterplot(data,x,y,category1,category2,title, xlabel, ylabel):
    sns.scatterplot(data = data, x = x, y = y, 
                    hue = category1, palette = ["#817dff", "#ffc07d"],
                    style = category2, markers = ["^", "v"])
    plt.legend(bbox_to_anchor=(1, 1), loc=2) ;
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()  


data = sns.load_dataset("tips")

medium_scatterplot(data,"tip","total_bill","sex","sex","Propina vs. Total pagado","Propina","Total pagado")
```

```python
# LINE PLOT
# BASIC

def basic_lineplot(data,x,y,title, xlabel, ylabel):
    plt.figure(figsize=(15,8))
    sns.lineplot(data = data, x = x, y = y, color = "green")
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()


data = pd.read_excel("tasas.xlsx", sheet_name = 'PERIODO_SUICIDIOSHOMBRES')
x = 'SUICIDIOS HOMBRES'
y = 'PERIODO'

basic_lineplot(data,'PERIODO','SUICIDIOS HOMBRES',"Total de suicidios de hombres por año","Año","Total suicidios hombres")
```

```python
# LINE PLOT
# MEDIUM

def medium_lineplot(data,x,y,category,title, xlabel,ylabel):
    plt.figure(figsize=(15,8))
    sns.lineplot(data = data, x = x, y = y,
                hue = category, palette = sns.color_palette("hls", 12))
    plt.legend(bbox_to_anchor=(1, 1), loc=2) ;
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()

data = sns.load_dataset("flights")

medium_lineplot(data,"year","passengers","month","Total de pasajeros por año","Año","Total pasajeros")
```

```python

# HISTOGRAM
# MEDIUM

def medium_histplot(data,x,category,title, xlabel):
    sns.histplot(data = data, x = x, 
                hue = category, palette = "rocket", multiple = "fill",
                binwidth = 500, binrange = [2500, 7000])
    #plt.legend(bbox_to_anchor=(1, 1), loc=2) ;
    plt.title(title)
    plt.xlabel(xlabel)
    plt.show()


data = sns.load_dataset("penguins")

medium_histplot(data,"body_mass_g","species","Masa corporal de los pingüinos", "Masa corporal")
```

```python
# BARRAS
# BASIC
data = sns.load_dataset("titanic")
x = "class"
y = "fare"
title = "Promedio pagado en función de la clase social"
xlabel = "Clase social"
ylabel = "Promedio pagado"

sns.barplot(data = data, x = x, y = y, palette = "Spectral",
	        order = ["Third", "Second", "First"], ci = None)
plt.title("Promedio pagado en función de la clase social")
plt.xlabel("Clase social")
plt.ylabel("Promedio pagado")
plt.show()
```

```python
# BARRAS
# MEDIUM

data = sns.load_dataset("titanic")
x = "class"
y = "fare"
title = "Promedio pagado en función de la clase social"
xlabel = "Clase social"
ylabel = "Promedio pagado"
hue = "sex"

def medium_barplot(data,x,y,hue):
    sns.barplot(data = data, x = x, y = y, 
                hue = hue, palette = "Spectral", ci = None)
    plt.legend(bbox_to_anchor=(1, 1), loc=2) ;
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()

medium_barplot(data,x,y,hue)
```
